Q. 다음은 파이썬 문법 중에서 False로 취급하는 것들 입니다.
앗, False로 취급하지 않는 것이 하나 있네요! True를 찾아주세요.

1)  None

2)  1

3)  ""

4)  0

5)  bool(0)

```python
values = [None, 1, "", 0, bool(0)]
results = {str(value): bool(value) for value in values}

for key, value in results.items():
  print(f"{key}: {value}")

# 정답 2번
```
