Q. 우리 태양계를 이루고 있는 행성은 수성, 금성, 지구, 화성, 목성, 토성, 천왕성, 해왕성으로 총 8개입니다. 저희는 우리 태양계의 n번째 행성이 무엇인지 알고 싶습니다.

입력으로 행성의 순서를 나타내는 숫자 n이 입력됩니다. 예를 들어 수성은 첫 번째 행성입니다.

```
>> 입력
1
```
```
>> 출력
수성
```

````python
n = int(input("몇 번째 행성을 출력할까요? (1-8): "))

planets = ["수성", "금성", "지구", "화성", "목성", "토성", "천왕성", "해왕성"]

print(f"{n}번째 행성은 {planets[n-1]}입니다.")
````
