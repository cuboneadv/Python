Q. 유주는 놀이공원 아르바이트 중입니다. 그런데 놀이기구마다 키 제한이 있습니다. 유주가 담당하는 놀이기구는 키가 150 이상만 탈 수 있습니다.

입력으로 키가 주어질 때, 키가 150 이상이면 **YES**를, 미만이면 **NO**를 출력하는 프로그램을 작성하세요.
```
>> 입력
190
```
```
>> 출력
YES
```
```
>> 입력
132
```
```
>> 출력
NO
```

````python
height = int(input("키를 입력하세요: "))

if height >= 150:
    print("YES")
else:
    print("NO")
````
