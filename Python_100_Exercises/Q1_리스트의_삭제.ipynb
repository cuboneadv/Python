Q. 다음 리스트에서 400, 500를 삭제하는 code를 입력하세요.

- nums = [100, 200, 300, 400, 500]

```python
# remove를 이용한 풀이
nums = [100, 200, 300, 400, 500]
nums.remove(400)
nums.remove(500)
print(nums)
```

```python
# pop을 이용한 풀이
nums = [100, 200, 300, 400, 500]
nums.pop()
nums.pop()
print(nums)
```

```python
# list comprehension을 이용한 풀이
nums = [100, 200, 300, 400, 500]
new_nums = [num for num in nums if num not in [400, 500]]
print(new_nums)
```
