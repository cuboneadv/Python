```
1) x = {1, 2, 3, 5, 6, 7}
```

## **set**
- 집합과 비슷한 특성을 지님
- 순서가 없고, set 안에서는 unique한 값을 가짐
- mutable 객체임
- 중괄호를 사용하여 dict와 비슷하지만, key가 없고 값만 존재함
- set 내부의 원소는 다양한 값을 함께 가질 수 있지만 mutable한 값은 가질 수 없음


- **immutable 객체**: 수정 불가능한 객체
> 정수, 실수, 문자열, 튜플 등
- **mutable 객체**: 수정 가능한 객체
> 리스트, 딕셔너리 등
