Q. 다음 중 변수 i가 6의 배수인지 확인하는 방법으로 올바른 것은?
```
1) i / 6 == 0
2) i % 6 == 0
3) i & 6 == 0
4) i | 6 == 0
5) i // 6 == 0
```

````python
# 변수 i가 6의 배수일 경우, i를 6으로 나누었을 때의 나머지는 0임
n = int(input("배수를 입력하세요: "))
i = int(input("변수 i를 입력하세요: "))

if i % 6 == 0:
    print(f"{i}은(는) {n}의 배수입니다.")
else:
    print(f"{i}은(는) {n}의 배수가 아닙니다.")

# 정답 2번
````
