Q. 우리 태양계를 이루는 행성은 수성, 금성, 지구, 화성, 목성, 토성, 천왕성, 해왕성이 있습니다. 이 행성들의 영어 이름은 Mercury, Venus, Earth, Mars, Jupiter, Saturn, Uranus, Neptune입니다.

**행성의 한글 이름을 입력하면 영어 이름을 반환하는 프로그램**을 만들어 주세요.
```
>> 입력
지구
```
```
>> 출력
Earth
```

````python
# 풀이1: if문 단순 나열을 통해
planet = input("행성의 한글 이름을 입력하세요: ")

if planet == "수성":
    print("Mercury")
elif planet == "금성":
    print("Venus")
elif planet == "지구":
    print("Earth")
elif planet == "화성":
    print("Mars")
elif planet == "목성":
    print("Jupiter")
elif planet == "토성":
    print("Saturn")
elif planet == "천왕성":
    print("Uranus")
elif planet == "해왕성":
    print("Neptune")
else:
    print("행성을 찾을 수 없습니다.")
````

````python
# 풀이2: dict를 통해
planet = input("행성의 한글 이름을 입력하세요: ")
planet_names = {
    "수성": "Mercury",
    "금성": "Venus",
    "지구": "Earth",
    "화성": "Mars",
    "목성": "Jupiter",
    "토성": "Saturn",
    "천왕성": "Uranus",
    "해왕성": "Neptune"
}

if planet in planet_names:
    print(planet_names[planet])
else:
    print("행성을 찾을 수 없습니다.")
````
