Q. **2-gram**이란 문자열에서 2개의 연속된 요소를 출력하는 방법입니다. 예를 들어 `Python`을 2-gram으로 반복한다면 결과는 다음과 같습니다.
```
>> 입력
Python
```
```
>> 출력
Py
yt
th
ho
on
```
입력으로 문자열이 주어지면 **2-gram**으로 출력하는 프로그램을 작성해주세요.

````python
s = input("문자열을 입력하세요: ")

def solution(s):
    s_len = len(s)
    for i in range(s_len-1):
        print(s[i:i+2])

print("\n2-gram으로 출력한 결과: ")
solution(s)
````
