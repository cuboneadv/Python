Q1. 알파벳 하나만을 입력하고 그 알파벳이 대문자이면 **YES**를, 아니면 **NO**를 출력하는 프로그램을 작성해주세요.

Q2. 알파벳 여러 개를 입력하고 입력한 알파벳 중 대문자인 알파벳만 출력하는 프로그램을 작성해주세요.

**isalpha() 메서드**
- 문자열이 알파벳 문자로만 구성되어 있는지를 확인하는 Python 문자열 메서드
- 문자열이 모두 알파벳 문자일 경우 True를 반환하고, 그렇지 않으면 False를 반환
- 알파벳 문자는 A-Z와 a-z로 정의
- 숫자, 공백, 특수 문자 등이 포함된 문자열은 False를 반환

```python
# Q1.
def solution(s):
    if not s.isalpha():
        return "알파벳을 정확하게 입력해주세요"
    elif s == s.upper():
        return "YES"
    else:
    # elif s == s.lower():
        return "NO"

alphabet = input("알파벳을 입력해주세요: ")

print(solution(alphabet))

# isalpha() 메서드를 이용하여 잘못된 입력값일 경우 경고 문구 출력
```

```python
# Q2.
def solution(s):
    upper_list = []
    for value in s:
        if not value.isalpha():
            return "알파벳을 정확하게 입력해주세요"
        if value.isupper():
            upper_list.append(value)
    return upper_list

alphabet = input("알파벳을 입력해주세요: ").split()

solution(alphabet)

# isupper() 메서드를 사용하여 입력값 중 대문자인 알파벳만 출력
```
