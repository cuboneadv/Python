Q. 문자 pineapple에서는 apple이라는 문자가 숨어있습니다. 재석이는 이와 같이 문자열 속에 숨어있는 문자를 찾아보려고 합니다.

입력으로 첫 번째 줄에 문자열이 주어지고 두 번째 줄에 찾을 문자가 주어지면, **그 문자가 시작하는 index**를 반환하는 프로그램을 만들어주세요.
```
>> 입력
pineapple is yummy
apple
```
```
>> 출력
4
```

**find() 메서드**
- 서브스트링 sub가 슬라이스 내에 등장하는 가장 작은 문자열의 인덱스를 돌려줌
- 선택적 인자 start 와 end 는 슬라이스 표기법으로 해석됨
- sub이 없으면 -1을 돌려줌
```
>> str.find(sub[,start[,end]])
```

````python
text = input("문자열을 입력하세요: ")
word = input("단어를 입력하세요: ")

index = text.find(word)

print(index)
````
