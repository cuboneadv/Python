Q. 반 친구들의 키가 주어지면, 친구들이 **키 순서대로** 올바르게 서 있는지 확인하는 프로그램을 작성해 주세요.
```
>> 입력
176 156 155 165 166 169
```
```
>> 출력
NO
```

````python
def solution(s):
    if s == sorted(s):
        return "YES"
    else:
        return "NO"

height = list(map(int, input("키를 입력해 주세요: ").split()))

print(solution(height))
````
