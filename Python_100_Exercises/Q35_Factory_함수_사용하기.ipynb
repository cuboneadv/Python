Q. 2제곱, 3제곱, 4제곱을 계산할 수 있는 팩토리 함수(Factory function)를 만들려고 합니다. `<pass>` 부분에 코드를 작성하여 `two` 함수를 완성하세요.
```
def one(n):
    def two(x):
        <pass>
    return two

a = one(2)
b = one(3)
c = one(4)

print(a(10))
print(b(10))
print(c(10))
```

- one(n) 함수는 내부에 two(x)라는 함수를 정의하고, 이 함수를 반환함
- 반환된 two(x) 함수는 n 값을 기억하며, x를 n 제곱하여 반환함
- a, b, c는 two 함수와 동일한 기능을 가짐

````python
def one(n):
    def two(x):
        return x ** n
    return two

a = one(2)
b = one(3)
c = one(4)

print(a(10))
print(b(10))
print(c(10))
````
