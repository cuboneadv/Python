Q. 호준이는 아르바이트로 영어 학원에서 단어 시험지를 채점하는 일을 하고 있다. 이 학원에서는 매번 1위부터 3위까지의 학생에게 상으로 사탕을 주고 있다. 그런데 오늘은 사탕이 다 떨어져서, 호준이가 채점한 후에 학생들의 점수를 바탕으로 사탕을 사러 가기로 했다.

학생들의 점수를 공백으로 구분하여 입력받으면, 1위부터 3위까지의 학생 숫자만큼 사탕을 사야 한다. 이때, 1위부터 3위까지의 학생들 중 중복된 학생도 포함하여 계산해야 한다.
```
>> 입력
97 86 75 66 55 97 85 97 97 95
```
```
>> 출력
6
```

````python
from collections import Counter

def count_candies(scores):

    score_count = Counter(scores)
    sorted_scores = sorted(score_count.keys())

    top_scores = sorted_scores[-3:]

    count = sum([score_count[score] for score in top_scores])

    return count

scores = list(map(int, input("점수를 입력하세요: ").split()))

candies_needed = count_candies(scores)

print(f"사탕을 {candies_needed}개 사야 합니다.")
````
