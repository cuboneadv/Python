Q. 테마파크에 온 경진이와 친구들이 놀이기구를 타려고 합니다. <br>
각 놀이기구는 무게 제한이 있으며, 이를 초과하면 다음 기구를 타야 합니다. <br>
주어진 무게 제한을 넘지 않도록 **최대 몇 명이 한꺼번에 탑승할 수 있는지 계산하는 프로그램을 작성해 주세요.**
- 첫 번째 줄에 놀이기구의 무게 제한이 주어집니다.
- 두 번째 줄에 친구들의 수 n이 주어집니다.
- 세 번째 줄부터 각 친구의 몸무게가 주어집니다.
```
>> 입력
50
5
20
20
20
20
20
```
```
>> 출력
2
```

````python
def solution(limit, n, weights):
    num = 0
    weights.sort()
    for i in range(n):
        if limit >= weights[i]:
            limit -= weights[i]
            num += 1
        else:
            break
    return num

limit = int(input("놀이기구의 무게 제한을 입력해 주세요: "))
n = int(input("친구들의 수를 입력해 주세요: "))
weights = list(map(int, input("친구들의 몸무게를 입력해 주세요: ").split()))

print(f"탑승 가능한 최대 인원은 {solution(limit, n, weights)}명 입니다.")
````
