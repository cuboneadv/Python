Q. 주어진 숫자가 소수인지 판별하는 프로그램을 작성해주세요.

소수일 경우 `YES`를 출력하고, 그렇지 않으면 `NO`를 출력해주세요."
```
>> 입력
11
```
```
>> 출력
YES
```
```
>> 입력
6
```
```
>> 출력
NO
```

````python
# 풀이1
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(11))
````

````python
# 풀이2
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    # 2를 제외한 짝수는 이미 False로 분류했으므로,
    # for문에서는 간격을 2로 설정하여 홀수만을 검사함
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

print(is_prime(1492871))
````
