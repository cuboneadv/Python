Q. 2020년 1월 1일은 수요일입니다. 주어진 a월 b일이 2020년의 어떤 요일인지 알아보는 함수를 작성하세요.

입력으로 a와 b가 주어지면, 2020년 a월 b일의 요일을 반환하는 함수를 작성해야 합니다. 요일은 일요일부터 토요일까지 각각 "SUN", "MON", "TUE", "WED", "THU", "FRI", "SAT"로 표기됩니다.

예를 들어, a = 5와 b = 24인 경우, 2020년 5월 24일은 일요일이므로 문자열 "SUN"을 반환해야 합니다.

**제한 조건**
- 2020년은 윤년입니다.
- 주어진 날짜는 실제로 존재하는 날짜입니다. (예: 13월 26일이나 2월 45일과 같은 날짜는 주어지지 않습니다.)

```python
# 풀이1: datetime 모듈 이용
from datetime import datetime

def solution(a, b):

    # 2020년 a월 b일 날짜 객체 생성
    date = datetime(2020, a, b)

    # 요일을 숫자로 가져옴 (0 ~ 6: 월요일 ~ 일요일)
    day_of_week = date.weekday()

    # 요일을 문자로 매핑
    days = ["MON", "TUE", "WED", "THU", "FRI", "SAT", "SUN"]

    # "수요일"인 경우, day_of_week는 2임
    # days[2]는 "WED"이므로, 숫자와 문자의 요일이 일치함
    answer = days[day_of_week]

    return answer

a = int(input("월을 입력하세요 : "))
b = int(input("일을 입력하세요 : "))

print(solution(a, b))
```

```python
# 풀이2: datetime 모듈 이용 X

def solution(a, b):

    # 각 월별 일수
    month_days = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    total_days = sum(month_days[:a-1]) + b

    # 1월 1일인 경우, total_days는 1임
    # 1 % 7 = 1이므로, days[1]이 "WED"가 되도록 조정
    days = ["TUE", "WED", "THU", "FRI", "SAT", "SUN", "MON"]

    answer = days[total_days % 7]

    return answer

a = int(input("월을 입력하세요 : "))
b = int(input("일을 입력하세요 : "))

print(solution(a, b))
```
