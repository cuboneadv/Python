Q. 다음은 숫자를 2진수로 변환하는 방법에 대한 설명입니다.

우리가 흔히 사용하는 숫자 1, 8, 19, 28893 등은 10진수 체계로 표현됩니다. 이 숫자들을 컴퓨터가 이해할 수 있는 2진수로 변환하려면 다음과 같은 방법을 사용합니다. 예를 들어, 13을 2진수로 변환하면 `2^3 + 2^2 + 2^0 = 13`이므로, 2진수로는 `1101`로 표현됩니다.

사용자가 입력한 숫자를 2진수로 변환하고 그 값을 출력하는 방법을 설명합니다. 여기서는 bin 함수를 사용하지 않고 변환 과정을 직접 구현합니다.
```
>> 입력
13
```
```
>> 출력
1101
```

````python
# 풀이1: for를 이용한 풀이

number_list = []
n = int(input())

# n을 2로 나눴을 때 몫이 0이라면 break
# for문의 조건에 정확한 반복 횟수를 지정하기 어렵다는 단점이 있음(임의의 큰 수인 '10000'을 사용)
for _ in range(10000):
    if n // 2 == 0:
        number_list.append(n % 2)
        break
    number_list.append(n % 2)
    n = n // 2

number_list.reverse()
number_list = list(map(str, number_list))

print(''.join(number_list))
````

````python
# 풀이2: while을 이용한 풀이

n = int(input())
number_list = []

while n:
    # print(n)
    number_list.append(n % 2)
    n = n // 2

number_list.reverse()
number_list = list(map(str, number_list))

print(''.join(number_list))
````
