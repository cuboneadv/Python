Q. **사용자가 입력한 양의 정수의 각 자리수 합을 구하는 프로그램**을 작성하세요.

예를 들어

18234의 경우, 각 자리수는 1, 8, 2, 3, 4이며, 이들의 합은 18입니다.
3849의 경우, 각 자리수는 3, 8, 4, 9이며, 이들의 합은 24입니다.

이와 같이 주어진 정수의 각 자리수를 더한 값을 출력하는 프로그램을 만들어 주세요.
```
>> 입력
1234
```
```
>> 출력
10
```

````python
# 풀이1
numbers = input("양의 정수를 입력하세요: ")

digits = list(numbers)

sum_digits = sum(int(digit) for digit in digits)

print(f"각 자리수의 합: {sum_digits}")
````

````python
# 풀이2: 입력 값이 양의 정수 형태인지 판별이 가능한 프로그램
input_number = input("양의 정수를 입력하세요: ")

# 입력 값이 양의 정수인지 검사
if input_number.isdigit():

    # 문자열에서 각 문자(자리수)를 정수로 변환하여 리스트로 생성
    digits = [int(digit) for digit in input_number]

    # 리스트에 있는 정수들의 합을 계산
    sum_of_digits = sum(digits)

    # 결과 출력
    print("각 자리수의 합:", sum_of_digits)

else:
    print("올바른 양의 정수를 입력하세요.")
````
