Q. Python의 **time 모듈**은 1970년 1월 1일 0시 0분 0초 이후로부터 현재까지 흐른 시간을 초 단위로 반환합니다.

이 기능을 사용하여 **현재 연도를 출력하는 프로그램**을 작성하세요.

```python
import time

current_time_seconds = time.time()
local_time = time.localtime(current_time_seconds)
current_year = local_time.tm_year

print(f"현재 연도: {current_year}")
```

```python
# 풀이1
import time

# 현재 시간의 초 단위 타임스탬프 획득
current_time_seconds = time.time()

# 타임스탬프를 날짜와 시간으로 변환
local_time = time.localtime(current_time_seconds)

# 연도, 월, 일, 시, 분, 초를 추출
current_year = local_time.tm_year
current_month = local_time.tm_mon
current_day = local_time.tm_mday
current_hour = local_time.tm_hour
current_minute = local_time.tm_min
current_second = local_time.tm_sec

# 결과 출력
print(f"현재 연도: {current_year}")
print(f"현재 날짜와 시간: {current_year}-{current_month:02d}-{current_day:02d} {current_hour:02d}:{current_minute:02d}:{current_second:02d}")
```

**pytz 라이브러리**([출처](https://wikidocs.net/236273))
- Python에서 시간대(time zone)를 처리하는 데 사용됨
- 이 라이브러리는 전 세계의 모든 시간대를 지원하며, 표준 시간대와 일광 절약 시간제를 다루는 데 유용함
- Python의 기본 datetime 모듈과 함께 사용하면 특정 시간대의 시간을 정확하게 관리하고 변환할 수 있음

```python
# 풀이2: 정확한 서울의 시간을 출력하기 위해, 다른 모듈 사용

from datetime import datetime
import pytz

# 서울 시간대 객체를 생성
seoul_tz = pytz.timezone('Asia/Seoul')

# 현재 서울 시간대의 시간을 획득
seoul_time = datetime.now(seoul_tz)

# 연도와 현재 시간을 출력
current_year = seoul_time.year
current_time = seoul_time.strftime("%Y-%m-%d %H:%M:%S")

print("현재 연도:", current_year)
print("현재 서울 시간:", current_time)
```
