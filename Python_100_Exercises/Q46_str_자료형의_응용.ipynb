Q. 1부터 20까지의 모든 숫자를 일렬로 나열하고, 이 숫자들의 모든 자리수의 합을 구하세요.

예를 들어, 10부터 15까지의 숫자를 일렬로 나열하면 101112131415가 되며, 각 자리수의 합은 25입니다.

````python
# 풀이1
number_list = []

for i in range(1, 21):
    number_list.append(str(i))

number_list = ''.join(number_list)

answer = 0
for i in number_list:
    answer += int(i)

print(answer)
````
