````python
def bubble(n, data):
    for i in range(n-1):
        # for j in range(n-1):
        for j in range(n-1-i): # j의 범위를 최소화하여 효율성 높이기
            if data[j] > data[j+1]:
                data[j], data[j+1] = data[j+1], data[j]
    for i in range(n):
        print(data[i], end = " ")

n = int(input())
data = list(map(int, input().split()))

bubble(n, data)
````
