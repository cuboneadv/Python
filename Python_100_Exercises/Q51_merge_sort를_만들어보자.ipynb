````python
# 풀이1
def 합병정렬(입력리스트):
    입력리스트길이 = len(입력리스트)
    if 입력리스트길이 <= 1:
        return 입력리스트

    중간값 = 입력리스트길이 // 2
    그룹_하나 = 합병정렬(입력리스트[:중간값])
    그룹_둘 = 합병정렬(입력리스트[중간값:])
    결과값 = []

    while (그룹_하나) and (그룹_둘):
        if (그룹_하나[0] <= 그룹_둘[0]):
            결과값.append(그룹_하나.pop(0))
        elif (그룹_하나[0] > 그룹_둘[0]):
            결과값.append(그룹_둘.pop(0))

    while 그룹_하나:
        결과값.append(그룹_하나.pop(0))
    while 그룹_둘:
        결과값.append(그룹_둘.pop(0))
    return 결과값

주어진리스트 = [180, 145, 165, 45, 170, 175, 173, 171]

print(합병정렬(주어진리스트))
````

````python
# 풀이2
def merge_sort(input_list):
    if len(input_list) <= 1:
        return input_list

    mid = len(input_list) // 2
    group1 = merge_sort(input_list[:mid])
    group2 = merge_sort(input_list[mid:])

    result = []
    index1 = 0
    index2 = 0

    while index1 < len(group1) and index2 < len(group2):
        if group1[index1] <= group2[index2]:
            result.append(group1[index1])
            index1 += 1
        else:
            result.append(group2[index2])
            index2 += 1

    result.extend(group1[index1:])
    result.extend(group2[index2:])
    return result

input_list = list(map(int, input("리스트의 값을 입력하세요: ").split()))
print(merge_sort(input_list))
````
