### 변수명으로 사용할 수 없는 것들

#### **예약어**
Python에서 변수명으로 사용할 수 없는 예약어들은 다음과 같음
* False, await, else, import, pass, None, break, except, in, raise,
True, class, finally, is, return, and, continue, for, lambda, try,
as, def, from, nonlocal, while, assert, del, global, not, with,
async, elif, if, or, yield 등<br><br>

#### **잘못된 형식의 변수명**
변수명을 정의할 때 다음 규칙을 따라야 함

* 변수명은 문자(A-Z, a-z), 숫자(0-9), 또는 밑줄(_)로 구성될 수 있음
* 변수명은 숫자로 시작할 수 없음
* 변수명은 대소문자를 구분함
* 변수명은 공백을 포함할 수 없음
* 변수명은 특수문자(!, @, #, $, %, ^, &, *, (, ), -, +, = 등)를 포함할 수 없음
